### Curvas


```r
# Para representar f(x)=x^2 entre -10 y 10 se usa
curve(x^2,-10,10) 
```

```r
curve(dnorm(x, mean = 100, sd = 20),xlim = c(10, 200))
```

```r
# Para agregar una leyenda

curve(dnorm(x, mean = 100, sd = 20),xlim = c(10, 200))
legend("topright",
       c(paste("mu =", 100), paste("SD =", 20)),
       bty = "n")
```

# Histograma

```r
datos <- c(6.9, 6.2, 5.4, 4.7, 4.8, 6.0, 6.3, 5.0, 4.5, 6.3,
       5.9, 6.7, 6.1, 5.0, 6.7, 5.8, 6.1, 5.7, 5.2, 4.6)

hist(datos)

hist(datos,col = "orange")
```

```r
hist(datos, col = "beige", main = "Mi histograma",
     xlab = "x", ylab = "y")
```

```r
muestra2 <- rnorm(10, 100, 20)
hist(muestra2, freq = F, main = "Mi histograma")
lines(density(muestra2), lty = 2)
```

# barplot

```r
datos <- c(20,48,106,3,23,33,120,67)
barplot(datos)
```

```r
# barplot pide como argumento una matriz, 
# que represente una tabla de contingencia con los datos a graficar. 
# Este tipo de tablas pueden ser generadas con la función table() . 
# table() pide como argumento uno o más vectores, de preferencia variables discretas.
# sample() es para obtener un aleatorio de nuestros
# prob() es la probabilidad de ocurrir .10 es 10%

barplot(table(sample(1, size=50, replace=TRUE, prob=c(.10))))
```

```r
barplot(table(sample(1:5, size=50, replace=TRUE, prob=c(.25,.50,.10,1,.80))))
```
